**Task 14 (2 points).** Find a team with the highest win rate in the **2015/2016** season. Win rate means won matches / all matches. If there are several teams with the highest win rate return the first by name in lexical order

```python
# Load match data and team data
match_data = pd.read_csv('match_data.csv', index_col="Unnamed: 0")
team_data = pd.read_csv('team_data.csv', index_col="Unnamed: 0")

# Filter matches for the 2015/2016 season
matches_2015_2016 = match_data[match_data['season'] == '2015/2016']

# Count the matches won by each team
home_wins = matches_2015_2016[matches_2015_2016['home_team_goal'] > matches_2015_2016['away_team_goal']]['home_team_api_id']
away_wins = matches_2015_2016[matches_2015_2016['away_team_goal'] > matches_2015_2016['home_team_goal']]['away_team_api_id']
won_matches = pd.concat([home_wins, away_wins])
won_matches_count = won_matches.value_counts()

# Count the total matches played by each team
total_matches = pd.concat([matches_2015_2016['home_team_api_id'], matches_2015_2016['away_team_api_id']])
total_matches_count = total_matches.value_counts()

# Calculate win rate for each team
win_rate = (won_matches_count / total_matches_count).fillna(0)

# Find the team with the highest win rate
team_with_highest_win_rate_id = win_rate[win_rate == win_rate.max()].sort_index().idxmax()

# Get the name of the team with the highest win rate
team_highest_winrate_2015_2016 = team_data[team_data['team_api_id'] == team_with_highest_win_rate_id]['team_long_name'].values[0]
```

**Task 15 (2 points).** Determine the team with the maximum days' gap between matches in **England Premier League 2010/2011 season**. Return number of days in that gap. <br>
**Note**: a *gap* means the number of days between two consecutive matches of the same team.

```python
# Load data
league_data = pd.read_csv("league_data.csv", index_col="Unnamed: 0")
match_data = pd.read_csv("match_data.csv", index_col="Unnamed: 0")

# Filter data for England Premier League 2010/2011 season
eng_premier_matches_2010_2011 = match_data[(match_data['league_id'] == 1729) & (match_data['season'] == '2010/2011')]

# Combine home and away matches for each team
home_matches = eng_premier_matches_2010_2011.groupby('home_team_api_id')['date'].apply(lambda x: pd.to_datetime(x).diff().max())
away_matches = eng_premier_matches_2010_2011.groupby('away_team_api_id')['date'].apply(lambda x: pd.to_datetime(x).diff().max())

# Concatenate home and away matches
all_matches = pd.concat([home_matches, away_matches]).reset_index()

# Get the team with the maximum gap between matches
team_with_max_gap_id = all_matches.loc[all_matches['date'].idxmax()]['index']

# Find the maximum gap in days
highest_gap_england_2010_2011 = all_matches['date'].max().days
```
